```python
import numpy as np; import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the predator-prey model
def predprey(t, y):
  return [1.2*y[0] - 0.6*y[0]*y[1], # Prey equation
  -0.8*y[1] + 0.3*y[0]*y[1]] # Predator equation

# Time span and initial conditions
t_span = (0, 20); y0 = [2, 1] # Initial population of prey and predator

# Solve the system of ODEs
solution = solve_ivp(predprey, t_span, y0, method='RK45', t_eval=np.linspace(0, 20, 300))
```

```python
# Plot the results
plt.figure(figsize=(12, 7)); plt.subplot(1, 2, 1)
plt.plot(solution.t, solution.y[0], label='Prey', markersize=3, marker = '+')
plt.plot(solution.t, solution.y[1], label='Predator', markersize=3, marker = '.')
plt.legend(); plt.title('Time plot'); plt.xlabel('Time'); plt.ylabel('Population')
plt.subplot(1, 2, 2); plt.plot(solution.y[0], solution.y[1])
plt.title('Phase plot'); plt.xlabel('Prey'); plt.ylabel('Predator')
```
