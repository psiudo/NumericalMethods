```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# 데이터 생성
np.random.seed(100)
x = np.linspace(-10, 10, 100)
y = np.sin(x)
x_tensor = torch.tensor(x, dtype=torch.float32).view(-1, 1)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# MLP 모델 정의
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.hidden = nn.Linear(1, 10)  # 은닉층
        self.output = nn.Linear(10, 1)  # 출력층
        self.activation = nn.Sigmoid()  # 활성화 함수

    def forward(self, x):
        x = self.activation(self.hidden(x))
        x = self.output(x)
        return x

# 모델 초기화
model = MLP()
criterion = nn.MSELoss()  # 손실 함수
optimizer = optim.Adam(model.parameters(), lr=0.01)  # 옵티마이저

# 학습
epochs = 5000
losses = []
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(x_tensor)
    loss = criterion(y_pred, y_tensor)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

    # 중간 결과 출력
    if epoch % 500 == 0 or epoch == epochs - 1:
        plt.scatter(x, y, label='True Data', color='blue')
        plt.plot(x, y_pred.detach().numpy(), label=f'Prediction (Epoch {epoch})', color='red')
        plt.legend()
        plt.title(f"Epoch {epoch} | Loss: {loss.item():.6f}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.show()

# 최종 결과 시각화
plt.figure()
plt.plot(range(epochs), losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.show()
```
