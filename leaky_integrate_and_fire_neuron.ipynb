```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters of the model
dt = 0.1  # integration time step [ms]
tau = 10  # time constant [ms]
ELeak = -65  # resting potential [mV]
theta = -55  # firing threshold [mV]
RIext = 12  # constant external input [Ohm*mA=mV]
```

```python
t = np.arange(0, 100 + dt , dt)
Vm = np.zeros(len(t)) # membrane potential
Vm[0] = ELeak
bSpike = np.zeros(len(t), dtype = bool)

for i in range (len(t) - 1):
  if Vm[i] > theta :
    Vm[i+1] = ELeak
    bSpike[i] = True
  else :
    Vm[i + 1] = Vm[i] + (dt / tau) * (ELeak - Vm[i] + RIext)
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_axes([0.13, 0.13, 1 - 0.26, 0.6])
ax1.plot(t, Vm, label='Vm (mV)')
ax1.plot([0, 100], [-55, -55], 'k--')
ax2 = fig.add_axes([0.13, 0.8, 1 - 0.26, 0.1])
ax2.plot(t, bSpike, '.', markersize=15, color='black')
ax2.set_ylim([0.5, 1.5]); plt.show()
```
