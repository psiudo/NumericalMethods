```python
import numpy as np
import matplotlib.pyplot as plt

# 정의하기
def f(t, y):
    return 4*np.exp(0.8 * t) - 0.5 * y

h = 1
t = np.arange(0, 4 + h, h)


y_euler  = np.zeros(len(t))
y_euler[0] = 2
y_heun = np.copy(y_euler)
y_midpoint = np.copy(y_euler)
y_rk4 = np.copy(y_euler)

# 수치해석
for i in range(len(t) - 1):
  # 오일러 방법
  y_euler[i + 1] = y_euler[i] + h * f(t[i], y_euler[i])

  # Heun의 방법
  y_predictor = y_heun[i] + f(t[i], y_heun[i]) * h
  mean_slope = ((f(t[i], y_heun[i])) + f(t[i+1], y_predictor))/2
  y_heun[i+1] = y_heun[i] + mean_slope * h

  # 중간점 방법
  y_predictor = y_midpoint[i] + f(t[i], y_midpoint[i]) * h/2
  y_midpoint[i+1] = y_midpoint[i] + f(t[i] + h/2, y_predictor)*h

  # Rungee - kunta 방법
  k1 = f(t[i], y_rk4[i])
  k2 = f(t[i] + h/2, y_rk4[i] + k1 * h /2)
  k3 = f(t[i] + h/2, y_rk4[i] + k2 * h /2)
  k4 = f(t[i] + h, y_rk4[i] + k3 * h)
  y_rk4[i+1] = y_rk4[i] + (k1 + 2*k2 + 2*k3 + k4) * h / 6


# 해석해
analytical_solution = (4 / 1.3) * (np.exp(0.8 * t) - np.exp(-0.5 * t)) + 2 * np.exp(-0.5 * t)
```

```python

# 그래프 그리기
plt.figure(1, figsize=(12, 10))

# Euler's method
plt.plot(t, y_euler, label="Euler's method", marker='o', linestyle='-', linewidth=0.5, markersize=6, color='blue')

# Analytical solution
plt.plot(t, analytical_solution, label="Analytical Solution", linestyle='--', marker='d', linewidth=0.5, markersize=6, color='green')

# Heun's method
plt.plot(t, y_heun, label="Heun's method", linestyle='--', marker='o', linewidth=0.5, markersize=6, color='red')

# Midpoint method
plt.plot(t, y_midpoint, label="Midpoint method", linestyle='--', marker='d', linewidth=0.5, markersize=6, color='purple')

# Rungee-Kunta method
plt.plot(t, y_rk4, label="Midpoint method", linestyle='--', marker='d', linewidth=0.5, markersize=6, color='yellow')


plt.xlabel('t')
plt.ylabel('y')
plt.title("Numerical Method vs Analytical Solution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 정의하기
def f(t, y):
    return y * t**3 - 1.5*y

h = 0.25
t = np.arange(0, 2 + h, h)


y_euler  = np.zeros(len(t))
y_euler[0] = 1
y_heun = np.copy(y_euler)
y_midpoint = np.copy(y_euler)
y_rk4 = np.copy(y_euler)

# 수치해석
for i in range(len(t) - 1):
  # 오일러 방법
  y_euler[i + 1] = y_euler[i] + h * f(t[i], y_euler[i])

  # Heun의 방법
  y_predictor = y_heun[i] + f(t[i], y_heun[i]) * h
  mean_slope = ((f(t[i], y_heun[i])) + f(t[i+1], y_predictor))/2
  y_heun[i+1] = y_heun[i] + mean_slope * h

  # 중간점 방법
  y_predictor = y_midpoint[i] + f(t[i], y_midpoint[i]) * h/2
  y_midpoint[i+1] = y_midpoint[i] + f(t[i] + h/2, y_predictor)*h

  # Rungee - kunta 방법
  k1 = f(t[i], y_rk4[i])
  k2 = f(t[i] + h/2, y_rk4[i] + k1 * h /2)
  k3 = f(t[i] + h/2, y_rk4[i] + k2 * h /2)
  k4 = f(t[i] + h, y_rk4[i] + k3 * h)
  y_rk4[i+1] = y_rk4[i] + (k1 + 2*k2 + 2*k3 + k4) * h / 6


# 해석해
analytical_solution = np.exp(0.25 * t**4 - 1.5*t)
```
