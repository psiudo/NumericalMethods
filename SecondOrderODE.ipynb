```python
import numpy as np
import matplotlib.pyplot as plt

# x(t) func def
def x(t):
    return (6 + 0.6 * t) * np.exp(-0.1 * t) + 29

# t range
t = np.arange(0, 150, 1)
x_values = x(t)

# plot
plt.plot(t, x_values, label=r'$x(t) = (6 + 0.6t)e^{-0.1t} + 29$')
plt.xlabel('t (ms)')
plt.ylabel('x(t)')
plt.title('Graph of $x(t)$ for $0 \leq t \leq 150$ ms')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation*
def odefun(t, y):
x, dx = y
ddx = 3 * t**2 - 5 * dx - 6 * x
return [dx, ddx]

# Initial conditions
t_span = (0, 2)
y0 = [1, 0]
t_eval = np.linspace(0, 2, 100)

# Solve numerically
sol = solve_ivp(odefun, t_span, y0, t_eval=t_eval)

# Calculate exact solution
t = t_eval
x_exact = (9/4) * np.exp(-2 * t) - (16/9) * np.exp(-3 * t) + 0.5 * t**2 - (5/6) * t + 19/36

# Plotting
plt.plot(sol.t, sol.y[0], 'b-', label='Numerical Solution')
plt.plot(t, x_exact, 'r--', label='analytic Solution')
plt.xlabel('Time t')
plt.ylabel('Solution x(t)')
plt.legend()
plt.title('Solution of the Differential equation')
plt.show()
```
