```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the differential equation system
def neuron_network(t, E, g=3):
    matrix = np.array([[-3, 0, -g, -5],
                       [-5, -3, 0, -g],
                       [-g, -5, -3, 0],
                       [0, -g, -5, -3]])
    return np.dot(matrix, E)

# Initial conditions
E0 = [1, 0, -1, 0]  # Example initial values for E1, E2, E3, E4
t_span = [0, 10]     # Time span from t=0 to t=10

# Solve the system using solve_ivp
sol = solve_ivp(neuron_network, t_span, E0, t_eval=np.linspace(0, 10, 100))

# Plot the results
plt.plot(sol.t, sol.y.T)
plt.xlabel('Time')
plt.ylabel('State Variables')
plt.title('Solution of the Neuron Network System')
plt.legend(['E1', 'E2', 'E3', 'E4'])
plt.show()
```
